```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_excel('IMDb Movies India.xlsx')

```

```python
df.head(100)
```

```python
df.shape
```

```python
columns = {'Actor 1','Actor 2','Actor 3','Genre','Director','Name'}
```

```python
df[columns].head(2)
```

```python
df[columns].isnull().values.any()
```

```python
def get_important_features(data):
  important_features = []
  for i in range(df.shape[0]):
    important_features.append(data['Actor 1'][i] +' ' + data['Actor 2'][i] + ' ' + data['Actor 3'][i] + ' ' + data['Genre'][i] + ' ' + data['Director'][i] + ' ' + data['Name'][i])
    if i == 2880:
      print(important_features[i])
    if i==2881:
      print(important_features[i])
  return important_features
```

```python
df['important_features'] = get_important_features(df)

df.head(3)
```

```python
cm = CountVectorizer().fit_transform(df['important_features'])
```

```python
print(cm)
```

```python
cs=cosine_similarity(cm)
print(cs[2880])
```

```python
cs.shape
```

```python
title = 'Swades'

movie_id = df[df.Name == title]['id'].values[0]
```

```python
print(movie_id)
```

```python
scores = list (enumerate(cs[movie_id]))
```

```python
print(scores)
```

```python
sorted_scores = sorted(scores, key = lambda x:x[1], reverse = True)
sorted_scores = sorted_scores[1:]
```

```python
print(sorted_scores)
```

```python
j=0

for item in sorted_scores:
  cvb = df[df.id == item[0]]['Name'].values[0]
  print(j+1, cvb, item[1])
  j=j+1
  if j>6:
    break
```
